```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Cargamos los datos limpios
df = pd.read_csv('./data/titanic_clean.csv')
```

```python
# Creamos variables 'dummy' para Sex y Embarked
# drop_first=True evita la redundancia (si no es hombre, es mujer)
df = pd.get_dummies(df, columns=['Sex', 'Embarked'], drop_first=True)

print("Columnas después del encoding:")
print(df.columns)
df.head()
```

```python
scaler = StandardScaler()

# Escalamos Age y Fare
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])

print("Datos escalados (ahora están en un rango similar):")
df.head()
```

```python
df.to_csv('./data/titanic_model_ready.csv', index=False)
print("¡Listo! Datos preparados para el entrenamiento en data/titanic_model_ready.csv")
```
