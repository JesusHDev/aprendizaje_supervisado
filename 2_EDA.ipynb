```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Cargamos el progreso del notebook anterior
df = pd.read_csv('./data/titanic_clean.csv')
df.head()
```

```python
# Gráfica de barras simple
plt.figure(figsize=(8, 5))
sns.countplot(x='Survived', data=df, palette='viridis')
plt.title('Conteo de Supervivencia (0 = No, 1 = Sí)')
plt.show()

# Tasa exacta
tasa_supervivencia = df['Survived'].mean() * 100
print(f"Tasa de supervivencia: {tasa_supervivencia:.2f}%")
```

```python
# ¿Influyó la clase social?
plt.figure(figsize=(10, 6))
sns.barplot(x='Pclass', y='Survived', hue='Sex', data=df)
plt.title('Supervivencia por Clase y Género')
plt.ylabel('Probabilidad de Supervivencia')
plt.show()
```
